## Import Section

```python
import requests
from bs4 import BeautifulSoup
```

## Define and get the HTML data

```python
url = "https://www.taiwanlottery.com.tw"
resp = requests.get(url)
```

```python
resp.status_code
```

```python
resp.text
```

## Specify BeautifulSoup Compiler

```python
soup = BeautifulSoup(resp.text, "lxml")
```

## Data Crawlings

```python
lottery_contents = soup.find("div", {"class": "contents_box02"})
lottery_contents
```

```python
lottery_title = lottery_contents.find("span", {"class": "font_black15"})
lottery_title
```

```python
lottery_title_text = lottery_title.text.strip()
lottery_title_text
```

```python
lottery_numbers = lottery_contents.find_all("div", {"class": "ball_tx ball_green"})
lottery_numbers
```

```python
lottery_number_list = [lottery_number.text.strip() for lottery_number in lottery_numbers]

print("Lottery Size:", len(lottery_numbers))
lottery_number_list
```

```python
lottery_red_number = lottery_contents.find("div", {"class": "ball_red"})
lottery_red_number
```

```python
lottery_red_number_text = lottery_red_number.text.strip()
lottery_red_number_text
```

## Output

```python
print("威力彩期數: ", lottery_title_text)
print("開出順序: ", "\t".join(lottery_number_list[:6]))
print("大小順序: ", "\t".join(lottery_number_list[6:]))
print("第二區: ", lottery_red_number_text)
```
